## 3. ***Data Wrangling***

### Data Wrangling Code

````python
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
import seaborn as sns
# Converting 'last_review' to datetime
Airbnb_df['last_review'] = pd.to_datetime(Airbnb_df['last_review'], errors='coerce')
````

````python
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
import seaborn as sns
# Handling outliers in 'price' and 'minimum_nights'

# Removing listings with a price greater than 99th percentile or less than 1st percentile
price_percentiles = Airbnb_df['price'].quantile([0.01, 0.99]).values
Airbnb_df = Airbnb_df[(Airbnb_df['price'] >= price_percentiles[0]) & (Airbnb_df['price'] <= price_percentiles[1])]

# Removing listings with minimum_nights greater than 99th percentile
min_nights_percentiles = Airbnb_df['minimum_nights'].quantile([0.01, 0.99]).values
Airbnb_df = Airbnb_df[(Airbnb_df['minimum_nights'] >= min_nights_percentiles[0]) & (Airbnb_df['minimum_nights'] <= min_nights_percentiles[1])]
````

````python
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
import seaborn as sns
# Creating a new feature - 'year_last_review' from 'last_review'
Airbnb_df['year_last_review'] = Airbnb_df['last_review'].dt.year.fillna(0).astype(int)
````

### Chart - 1

````python
# Checking the distribution of prices for the Airbnbs in NYC

# Set Seaborn theme for a clean look
sns.set_theme(style='darkgrid')

# Create the histogram with KDE overlay
plt.figure(figsize=(12, 6))
sns.histplot(
    Airbnb_df['price'],
    bins=50,
    kde=True,
    color='r',
    edgecolor='black',
    alpha=0.7
)

# Add title and axis labels
plt.title('Distribution of Airbnb Prices in NYC', fontsize=16, fontweight='bold')
plt.xlabel('Price (USD)', fontsize=14)
plt.ylabel('Density', fontsize=14)

# Set x-axis limit to focus on relevant price ranges (exclude extreme outliers)
plt.xlim(0, Airbnb_df['price'].quantile(0.95))  # Focus on prices up to the 95th percentile

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().set_facecolor("#f7f7f7")

# Show the plot
plt.tight_layout()
plt.show()
````

#### 1. Why did you pick the specific chart?



* The histogram shows the actual count, while the KDE overlay smoothens the data to reveal trends.

* Limiting the x-axis removes the influence of extreme outliers, making the chart more informative.

#### 2. What is/are the insight(s) found from the chart?


* A significant peak near lower price ranges (e.g., $50–$150) suggests most listings are affordable and cater to budget-conscious travelers.

### Chart - 2

````python
# Checking the distribution of Room Types

# Set up the figure
plt.figure(figsize=(12, 8))

# Get room type counts
room_type_counts = Airbnb_df['room_type'].value_counts()
labels = room_type_counts.index
sizes = room_type_counts.values

# Define colors for the pie chart
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Create the pie chart
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,  # Start angle for better layout
    colors=colors,
    explode=[0.05 if i == max(sizes) else 0 for i in sizes],  # Slightly explode the largest segment
    textprops={'fontsize': 12}  # Increase text size for readability
)

# Add a title
plt.title('Room Type Distribution', fontsize=16, fontweight='bold')

# Add a legend for better readability
plt.legend(
    title='Room Type',
    loc='upper left',
    fontsize=12,
    bbox_to_anchor=(1, 0.5),
    frameon=False
)

# Display the chart
plt.tight_layout()
plt.show()
````

#### 1. Why did you pick the specific chart?

* A pie chart effectively represents proportions, making it ideal for visualizing the distribution of room types.

* Including percentages and using an exploded segment highlights the most popular category for easy identification.
